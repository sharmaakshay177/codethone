# Seaborn For Visualization

```python
# Distribution plots in seaborn 
# to the check the distribution of the data set
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# distribution plot
sns.distplot(tips['total_bill'])
```

```python
#this is used in case you want to compare distribution of two columns 
# in a data set
sns.jointplot(x = 'total_bill',y='tip',data=tips,kind='reg')
#extra param kind = 'reg','hex','kde'and by default its scatter 
```

```python
tips.head()
```

```python
# pair plot will give pair wise relationship of every numerical entity
# in the given data set
sns.pairplot(tips)
```

```python
# in the above pairplot if we want to give categorical data
# there is param called hue= 'column name' and it will take that 
# column name 
sns.pairplot(tips,hue = 'sex')

# palette will take the color scheme 
```

```python
# to check the distribution line of the column 
# we can also draw a rug plot 
sns.rugplot(tips['total_bill'])
```

```python
# for avg data for a categorical column with numerical one 
sns.barplot(x='sex',y='total_bill',data= tips)

# estimator param is the one which will take the aggregate function if you want 
# to specify any by default its mean/avg
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data= tips)
```

```python
sns.violinplot(x='day',y='total_bill',data= tips)
```

```python
sns.violinplot(x='day',y='total_bill',data= tips, hue='sex',split=True)
```
